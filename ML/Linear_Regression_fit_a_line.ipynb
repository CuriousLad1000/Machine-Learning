**Linear Regression**

This code implements the derived calculations to create a linear solver. The code creates an empty canvas and user can click on it to add points. Right click on the canvas to draw a regression line moving through those points.





note: it seems I forgot to add 1/n in equations... but since I directly took the mean instead of sum and then 1/n, it should be fine...

```python
%matplotlib tk                   
import matplotlib.pylab as plt
import numpy as np
from matplotlib.backend_bases import MouseButton

f,a = plt.subplots() 

x1 = np.linspace(-2, 5, num=20) #will create an array of 20 numbers between -2 to 5. used to create fixed empty canvas (by drawing invisible line)
y1 = np.linspace(-2, 5, num=20) #will create an array of 20 numbers between -2 to 5. used to create fixed empty canvas (by drawing invisible line)

a.plot(x1,y1,alpha=0)           #Draws invisible diagonal line to lock dimensions of canvas.
posX = []                       #List to store X coordinates of mouse
posY = []                       #List to store Y coordinates of mouse


#Linearsolver
def my_linfit(x,y):            #Function to fit the line
    mean_x = np.mean(x)        #calculates mean of all x coordinates
    mean_y = np.mean(y)        #calculates mean of all y coordinates
    mean_xy = np.mean(x*y)     #calculates mean of all x*y coordinates
    mean_xx = np.mean(x*x)     #calculates mean of all x*x (x square's mean)
    a = ((mean_xy)-(mean_x*mean_y))/((mean_xx)-(mean_x*mean_x))        #Derived equation in terms of a
    b = ((mean_xx*mean_y)-(mean_x*mean_xy))/(mean_xx-(mean_x*mean_x))  #Derived equation in terms of b
    #b = mean_y -(a*(mean_x))                                          #Shortcut - find a and put in b let computer do the rest!
    
    return a,b                   #Returns a,b back

def on_click(event):
    global X_arr         #makes X_arr accessible by other iterations and to the rest of the program
    global Y_arr         #makes Y_arr accessible by other iterations and to the rest of the program
    global xp            #makes xp accessible by other iteraations and to the rest of the program
    if event.button == MouseButton.LEFT:       #checks for the left click and triggers actions
        posX.append([event.xdata])             #append current X coordinates to list
        posY.append([event.ydata])             #append current Y coordinates to list
        X_arr=np.array(posX, dtype = float)    #Converts list to array and stores in X_arr 
        Y_arr=np.array(posY, dtype = float)    #Converts list to array and stores in Y_arr
        
        #print(f"X pos: inx={len(posX)} and X={posX} and sum:{np.sum(X_arr)}") 
        #print(f"Y pos: inx={len(posY)} and Y={posY} and sum:{np.sum(Y_arr)}")
        
        plt.plot(event.xdata,event.ydata,'kx') #plots the x mark on screen at current coordinate
        plt.show()                             #show plot
    else: #on any other click (right or middle)
        a,b = my_linfit(X_arr,Y_arr)     #Run function to calculate fitline     
        
        xp = np.arange(-2,5,0.1)           #Create a line to fit with the calculated a and b variables
        plt.plot(xp,a*xp+b,'r-')           #plot regressed line
        print(f"My fit: a={a} and b={b}")  #print a and b values
        plt.show()                         #show plot
       
       # plt.disconnect(binding_id)   #enable this to stop interation with the graph after first regressed line.
       
        
binding_id = f.canvas.mpl_connect('button_press_event', on_click) #captures click events
```
