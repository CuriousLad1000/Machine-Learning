<b>Split data into batches Data Pre-processing</b>

In this code we will split the data into smaller batches that are balanced using train test split and stratify. The code uses quick and dirty method to split the data. The data is assumed to be aready processed, applied SMOTE, removed dupes, formatted in 8x8x3.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from numpy import load, savez_compressed
import gc
import cv2
```

***Now, LOAD DATA***

```python
# load dict of arrays
dict_data_X = load('../input/processed-total-dataset/Processed_X.npz')
dict_data_Y = load('../input/processed-total-dataset/Processed_Y.npz')


# extract the first array
X = dict_data_X['arr_0']
Y = dict_data_Y['arr_0']
# print the array
print("Training Dataset's Dimensions:" ,X.shape,"class Dim: ", Y.shape)
print()

del dict_data_X
del dict_data_Y


#scale values between 0 and 1

X = (X - np.min(X)) / (np.max(X) - np.min(X))
X = np.array(X).astype(np.float32)  
#Y = np.array(Y).astype(np.float32) 

gc.collect()
```

```python
#Plot first image
plt.figure(1);
plt.clf()
plt.imshow(X[0], cmap='gray')
plt.pause(1)
```

***Split data in 100,000 image batches (13 batches total)***

***100 images per class total 1000 classes***

```python
X1, Batch1_images, Y1, Batch1_Labels = train_test_split(X, Y, test_size=100000, train_size=1200000, random_state=42, stratify=Y)
print("Batch 1 Size:", Batch1_images.shape," Remaining:", X1.shape)

X1, Batch2_images, Y1, Batch2_Labels = train_test_split(X1, Y1, test_size=100000, train_size=1100000, random_state=42, stratify=Y1)
print("Batch 2 Size:", Batch2_images.shape," Remaining:", X1.shape)

X1, Batch3_images, Y1, Batch3_Labels = train_test_split(X1, Y1, test_size=100000, train_size=1000000, random_state=42, stratify=Y1)
print("Batch 3 Size:", Batch3_images.shape," Remaining:", X1.shape)

X1, Batch4_images, Y1, Batch4_Labels = train_test_split(X1, Y1, test_size=100000, train_size=900000, random_state=42, stratify=Y1)
print("Batch 4 Size:", Batch4_images.shape," Remaining:", X1.shape)

X1, Batch5_images, Y1, Batch5_Labels = train_test_split(X1, Y1, test_size=100000, train_size=800000, random_state=42, stratify=Y1)
print("Batch 5 Size:", Batch5_images.shape," Remaining:", X1.shape)

X1, Batch6_images, Y1, Batch6_Labels = train_test_split(X1, Y1, test_size=100000, train_size=700000, random_state=42, stratify=Y1)
print("Batch 6 Size:", Batch6_images.shape," Remaining:", X1.shape)

X1, Batch7_images, Y1, Batch7_Labels = train_test_split(X1, Y1, test_size=100000, train_size=600000, random_state=42, stratify=Y1)
print("Batch 7 Size:", Batch7_images.shape," Remaining:", X1.shape)

X1, Batch8_images, Y1, Batch8_Labels = train_test_split(X1, Y1, test_size=100000, train_size=500000, random_state=42, stratify=Y1)
print("Batch 8 Size:", Batch8_images.shape," Remaining:", X1.shape)

X1, Batch9_images, Y1, Batch9_Labels = train_test_split(X1, Y1, test_size=100000, train_size=400000, random_state=42, stratify=Y1)
print("Batch 9 Size:", Batch9_images.shape," Remaining:", X1.shape)

X1, Batch10_images, Y1, Batch10_Labels = train_test_split(X1, Y1, test_size=100000, train_size=300000, random_state=42, stratify=Y1)
print("Batch 10 Size:", Batch10_images.shape," Remaining:", X1.shape)

X1, Batch11_images, Y1, Batch11_Labels = train_test_split(X1, Y1, test_size=100000, train_size=200000, random_state=42, stratify=Y1)
print("Batch 11 Size:", Batch11_images.shape," Remaining:", X1.shape)

Batch13_images, Batch12_images, Batch13_Labels, Batch12_Labels = train_test_split(X1, Y1, test_size=100000, train_size=100000, random_state=42, stratify=Y1)
print("Batch 12 Size:", Batch12_images.shape," Remaining:", Batch13_images.shape)
```

```python
##To Save file
savez_compressed('Batch1_Y.npz', Batch1_Labels) 
savez_compressed('Batch1_X.npz', Batch1_images)  

savez_compressed('Batch2_Y.npz', Batch2_Labels) 
savez_compressed('Batch2_X.npz', Batch2_images)  

savez_compressed('Batch3_Y.npz', Batch3_Labels) 
savez_compressed('Batch3_X.npz', Batch3_images)  

savez_compressed('Batch4_Y.npz', Batch4_Labels) 
savez_compressed('Batch4_X.npz', Batch4_images)  

savez_compressed('Batch5_Y.npz', Batch5_Labels) 
savez_compressed('Batch5_X.npz', Batch5_images)  

savez_compressed('Batch6_Y.npz', Batch6_Labels)
savez_compressed('Batch6_X.npz', Batch6_images)  

savez_compressed('Batch7_Y.npz', Batch7_Labels) 
savez_compressed('Batch7_X.npz', Batch7_images)  

savez_compressed('Batch8_Y.npz', Batch8_Labels)
savez_compressed('Batch8_X.npz', Batch8_images) 

savez_compressed('Batch9_Y.npz', Batch9_Labels) 
savez_compressed('Batch9_X.npz', Batch9_images)  

savez_compressed('Batch10_Y.npz', Batch10_Labels) 
savez_compressed('Batch10_X.npz', Batch10_images)  

savez_compressed('Batch11_Y.npz', Batch11_Labels) 
savez_compressed('Batch11_X.npz', Batch11_images)  

savez_compressed('Batch12_Y.npz', Batch12_Labels) 
savez_compressed('Batch12_X.npz', Batch12_images) 

savez_compressed('Batch13_Y.npz', Batch13_Labels) 
savez_compressed('Batch13_X.npz', Batch13_images) 

```
